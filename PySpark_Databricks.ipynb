Actual Preprocessing

```
import numpy as np
def func1(x):
    try:
        x= int(x)
        return x
    except:
        return np.nan
    
def func2(x):
    try:
        x= float(x)
        return x
    except:
        return np.nan
```

```
from pyspark.sql.functions import udf, col
from pyspark.sql.types import FloatType, IntegerType

udf_int = udf(lambda x:func1(x), IntegerType() )
udf_flt = udf(lambda x:func2(x), FloatType() )
```

```
df= df.withColumn("MPG_cl", udf_int(col("MPG")))
df= df.withColumn("DISPLACEMENT_cl", udf_flt(col("DISPLACEMENT")))
df= df.withColumn("HORSEPOWER_cl", udf_int(col("HORSEPOWER")))
df= df.withColumn("WEIGHT", udf_int(col("WEIGHT")))
```
